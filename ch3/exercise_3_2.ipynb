```python
import sympy as sp
```

```python
y = sp.Function('y')
x = sp.symbols('x')
x, y(x)
```

# (a)

```python
diff_eq = sp.Eq(x*y(x).diff(x,2) - (x + 10)*y(x).diff(x) + 10*y(x),0)
diff_eq
```

```python
diff_eq.subs({y(x):sp.exp(x)}).simplify()
```
