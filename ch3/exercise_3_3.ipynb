```python
import sympy as sp
```

# Q 65.

```python
m = sp.symbols('m')
aux_eq65 = sp.Eq(m**3 - 6*m**2 + 2*m + 1, 0)
aux_eq65
```

```python
roots = sp.solve(aux_eq65)
roots
```

```python
for r in roots:
    print(sp.N(r, 5))  # 5 significant figures
```

# Q. 69

```python
m = sp.symbols('m')
aux_eq69 = sp.Eq(2*m**4 + 3*m**3 - 16 * m**2 + 15 * m - 4, 0)
aux_eq69
```

```python
sp.roots(aux_eq69)
```

```python
c1, c2, c3, c4 = sp.symbols('c_1 c_2 c_3 c_4')
x = sp.symbols('x')
gen_sol = c1 * sp.exp(-4*x) + c2 * sp.exp(1/2*x) + c3 * sp.exp(1*x) + c4 * x * sp.exp(1*x)
gen_sol
```

```python
y = gen_sol
y
```

```python
eq1 = sp.Eq(y.subs({x:0}), -2)
eq1
```

```python
eq2 = sp.Eq(sp.diff(y, x).subs({x:0}), 6)
eq2
```

```python
eq3 = sp.Eq(sp.diff(y, x, 2).subs({x:0}), 3)
eq3
```

```python
eq4 = sp.Eq(sp.diff(y, x, 3).subs({x:0}), 1/2)
eq4
```

```python
sp.solve([eq1, eq2, eq3, eq4], [c1, c2, c3, c4])
```

# Q. 70

```python
aux_eq70 = sp.Eq(m**4 - 3*m**3 + 3*m**2 - m, 0)
aux_eq70
```

```python
sp.roots(aux_eq70)
```

```python
gen_sol = c1 + c2*sp.exp(x*1) + c3*x*sp.exp(x*1) + c4*(x**2)*sp.exp(x*1)
gen_sol
```

```python
eq1 = sp.Eq(gen_sol.subs({x:0}), 0)
eq1
```

```python
eq2 = sp.Eq(sp.diff(gen_sol, x).subs({x:0}), 0)
eq2
```

```python
eq3 = sp.Eq(sp.diff(gen_sol, x, 2).subs({x:0}), 1)
eq3
```

```python
eq4 = sp.Eq(sp.diff(gen_sol, x, 3).subs({x:0}), 1)
eq4
```

```python
sp.solve([eq1, eq2, eq3, eq4], [c1, c2, c3, c4])
```
