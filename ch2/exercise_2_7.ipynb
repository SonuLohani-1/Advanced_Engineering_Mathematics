# Q. 51

## (a)

```python
t, v0, g = symbols('t v_0 g', positive=True)

# velocity and position from Problem 36
v = v0 - g*t
s = v0*t - Rational(1,2)*g*t**2
```

```python
# Time to reach max height (v = 0)
t_a = solve(Eq(v,0), t)
t_a
```

```python
# Time to fall back to ground (s = 0, t > 0)
t_total = solve(Eq(s,0), t) # second root is nonzero one
t_d = t_total[0] - t_a[0]        # descending portion
t_d
```

```python
t_d == t_a[0]
```

```python
# Impact velocity: substitute t_total back into v(t)
v_impact = v.subs(t, t_total[0])
simplify(Abs(v_impact) - v0) # This should be 0 or nearly zero
```

## (b)

```python
k, m = symbols('k m', positive=True)

# From Problem 37, with drag proportional to velocity
# The velocity and position functions are:
v_air = (v0 + m*g/k)*exp(-k*t/m) - m*g/k
s_air = (v0 + m*g/k)*(m/k)*(1 - exp(-k*t/m)) - (m*g/k)*t

# Substitute known values: weight=16lb → m=16/32=0.5 slug,
# g=32, v0=300, k=0.0025
vals = {m: Rational(1,2), g: 32, v0: 300, k: Rational(25,10000)}
```

```python
# t_a: solve v_air = 0  (still solvable analytically here)
t_a_air = solve(v_air.subs(vals), t)[0]
t_a_air.evalf()
```

```python
# t_d: solve s_air = 0 numerically 
s_air_num = s_air.subs(vals)
t_ground = nsolve(s_air_num, t, 30)  # 30 is initial guess
t_d_air = t_ground - t_a_air
t_d_air.evalf()
```

```python
# Compare t_a vs t_d, and impact speed vs v0
v_impact_air = v_air.subs(vals).subs(t, t_ground)
v_impact_air
```
