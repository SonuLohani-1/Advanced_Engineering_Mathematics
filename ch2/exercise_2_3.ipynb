```python
from sympy import *
from scipy.optimize import fsolve, minimize
```

# Q. 54

```python
x = symbols('x')
y = symbols('y', cls=Function)
diff_eq = Eq(y(x).diff(x) - 2*x*y(x), 1)
diff_eq
```

```python
sol = dsolve(diff_eq, ics={y(0):sqrt(pi)/2})
sol
```

```python
sol_at_2 = sol.rhs.subs(x, 2)
print(sol_at_2.evalf())
```

# Q. 55

```python
x = symbols('x')
y = Function('y')
diff_eq = Eq(x**3 * y(x).diff(x) + 2*x**2*y(x), 10*sin(x))
diff_eq
```

```python
sol = dsolve(diff_eq, ics={y(1):0})
sol
```

```python
y_expr = sol.rhs
y_func = lambdify(x, y_expr, modules=['scipy'])
```

```python
result = minimize(lambda x: -y_func(x), x0=1.0)  # x0 is initial guess

print(f"Maximum at x = {result.x[0]:.6f}")
print(f"Maximum value = {-result.fun:.6f}")
```
