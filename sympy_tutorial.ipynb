## Power

```python
x, y = symbols('x y', positive=True)
a, b = symbols('a b', real=True)
z, t, c = symbols('z t c')
```

```python
powsimp(x**a*x**b)
```

```python
# powsimp doesn't simplify if the identity is not valid for e.g
powsimp(z**c*t**c)
```

```python
# but we can force the simplification
powsimp(t**c*z**c, force=True)
```

```python
expand_power_exp(x**(a+b))
```

```python
expand_power_base((x*y)**a)
```

```python
# again it doesn't simplify it is not sure about the validity of the identity
expand_power_base((z*t)**c)
```

```python
def list_to_frac(l):
    expr = Integer(0)
    for i in reversed(l[1:]):
        expr += i
        expr = 1/expr
    return l[0] + expr
```

```python
list_to_frac([x, y, z])
```

```python
list_to_frac([1, 2, 3, 4, 5, 6, 7])
```

```python
syms = symbols('a0:5')
frac = list_to_frac(syms)
frac
```
