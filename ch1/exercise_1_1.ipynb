```python
from sympy import *
```

# Q. 63

```python
x = symbols('x')
y = symbols('y', cls=Function)
y(x)
```

```python
diffeq = Eq(y(x).diff(x,4) - 20*y(x).diff(x, 3) + 158*y(x).diff(x, 2) -580*y(x).diff(x) + 841*y(x), 0)
diffeq
```

```python
func = x*exp(5*x)*cos(2*x)
func
```

```python
diffeq.subs(y(x), func).simplify()
```

# Q. 64

```python
diffeq = Eq(x**3*y(x).diff(x, 3) + 2*x**2*y(x).diff(x, 2) + 20*x*y(x).diff(x) - 78*y(x), 0)
diffeq
```

```python
func = 20*cos(5*ln(x))/x - 3*sin(5*ln(x))/x
func
```

```python
diffeq.subs(y(x), func).simplify()
```
